# 1. CREATING PLOTS USING OBJECT ORIENTED STYLE

# simple plot

```python
x = np.arange(1,7)
y = np.arange(7,13)
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y,marker='*',markersize=10,c ='b');
ax.set_title('X aagainst Y',fontsize=15,c='r')
ax.set_xlabel('X-axis',fontsize=16,c='r')
ax.set_ylabel('Y-axis',fontsize=16,c='r')
plt.xticks(fontsize=14);
plt.yticks(fontsize=14);
ax.grid(True)
plt.style.use('classic')

```

# histogram

```python
x = np.random.randn(150)
fig,ax=plt.subplots()
ax.hist(x,bins=10,color='tomato',density=1);
plt.style.use('classic')
ax.grid(True)
ax.set_title('A Simple Histogram Of Random Numbers',fontsize=16,color='b')
ax.set_xlabel('X-Axis',fontsize=15,color='b')
ax.set_ylabel('Y-Axis',fontsize=15,color='b')
plt.show()
```

# 2.CREATING A DATAFRAME WITH NUMERIC AND CATEGORICAL DATATYPES USING PANDAS AND THEN CREATING A HISTOGRAM AND BAR CHART FROM IT

```python
"""The data created here are not authentic
they are just for illustration"""

data = {'Names':['Israel','Femi','Aminah','Emma','Faidah','Edet','Adeola','Stephen','Kafee','Edet','Edet','Israel','Faidah'],
      'Sex':['M','M','F','M','F','M','F','M','F','M','M','M','F'],'Bmi':[33.5,25.6,30.5,27.1,33.1,28.9,37.6,23.9,25.0,22.2,19.0,24.7,23],
      'Age':[25,26,27,28,30,15,26,25,33,24,16,22,36]}
df = pd.DataFrame(data)
df
```

```python
# Creating a histogram with the categorical values
x = df.Names
fig,ax = plt.subplots()
ax.hist(x,color='g');
ax.set_title('Simple histogram showing sex distribution',fontsize=16,color='m')
ax.set_xlabel('Sex',fontsize=15,color='m')
ax.grid(True)
plt.style.use('classic')
plt.show()
```

```python
#Creating a Barchart from the numerical values
x = df.Names
y = df.Bmi
fig,ax = plt.subplots()
ax.bar(x,y,color='c')
ax.set_xlabel('Names',color='g')
ax.set_ylabel('Bmi',color='g')
plt.show()
```

# Creating a Pie Chart

```python
labels = ['Data-Science', 'Web-development', 'Graphic-design','App-development']
sizes = [66.5, 30.6, 44.7, 59.3]
colors = ['yellow', 'blue', 'red', 'cyan']
explode = (0.1,0,0,0)
fig,ax = plt.subplots()
ax.pie(sizes, colors=colors, shadow=True, startangle=90,labels=labels,
       explode=explode,autopct='%1.1f%%')
ax.axis('equal')
plt.tight_layout()
plt.show()
```
